```python
from geopy.geocoders import Nominatim

import pandas as pd
geolocator = Nominatim(user_agent="geoapi")

def get_district(lat, lon):
    try:
        location = geolocator.reverse((lat, lon), exactly_one=True, language='en')
        address = location.raw.get('address', {})
        return address.get('suburb', address.get('city', 'Unknown'))  # Try 'suburb', fallback to 'city'
    except Exception as e:
        return 'Unknown'

df['district'] = df.apply(lambda x: get_district(x['latitud'], x['longitud']), axis=1)

# heatmap of the districts
import folium
from folium.plugins import HeatMap

m = folium.Map(location=[-34.61, -58.44], zoom_start=12)

heat_df = df[['latitud', 'longitud']]
heat_df = heat_df.dropna(axis=0, subset=['latitud','longitud'])

heat_data = [[row['latitud'],row['longitud']] for index, row in heat_df.iterrows()]

HeatMap(heat_data).add_to(m)

m.save('Data/heatmap.html')

```
